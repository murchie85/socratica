# MULTI LAMBDA INPUT
## combine first and last name & capitalize


```python
full_name = lambda fn, ln: fn.strip().title() + " " + ln.strip().title()
```

```python
## Note we strip space, and capitalize 
full_name('adam', 'mcmurchie')
```

```python
full_name('  adam', 'Mcmurchie')
```

#  LAMBDA WITHOUT A NAME

```python
scifi_authors = ["Isaac Asimov", "Ray Bradbury", "Arthur C. Clarke", "Douglas Adams", "H. G. Wells", "Robert Heinlein"]
```

```python
#Lets sort by last name 
# Note some have a middle name 
# Note some have initials
```

```python
# KEY is action for sorting, we will pass in lambda
help(scifi_authors.sort)
```

```python
scifi_authors.sort(key=lambda name: name.split(" ")[-1].lower())
scifi_authors
```

```python
def build_quadratic_function(a,b,c):
    """Returns the function f(x) = ax^2 + bx + c"""
    return lambda x: a*x**2 + b*x + c


f = build_quadratic_function(2, 3, -5)
```

```python
f(0)
```

```python
f(1)
```

```python
f(2)
```

```python
build_quadratic_function(3,0,1)
```

```python
build_quadratic_function(3,0,1)(2)
```
